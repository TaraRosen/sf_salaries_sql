```python
# import libraries
import pandas as pd
import sqlite3
```

```python
# connect to database
conn = sqlite3.connect('''database.sqlite''')

# create cursor object
cur = conn.cursor()
```

```python
cur.execute('''SELECT * FROM Salaries;''')               # sql query
salaries_df = pd.DataFrame(cur.fetchall())               # create dataframe from sql query
salaries_df.columns = [x[0] for x in cur.description]    # labels dataframe columns
salaries_df                                              # view dataframe
```

```python
# The result of a "cursor.execute" can be iterated over by row
for row in cur.execute('SELECT * FROM Salaries;'):
    print(row)
```
